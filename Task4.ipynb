```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create dummy "gesture-like" data (for demo)
# 0 = Open Hand, 1 = Closed Fist
X = [
    [1, 1], [1, 2], [2, 1],   # Open hand
    [5, 5], [6, 5], [5, 6]    # Closed fist
]

y = [0, 0, 0, 1, 1, 1]

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X, y)

# Prediction
pred = model.predict([[2,2]])
print("Prediction (0=Open Hand, 1=Closed Fist):", pred[0])

# Plot
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='green', label='Open Hand' if i==0 else "")
    else:
        plt.scatter(X[i][0], X[i][1], color='red', label='Closed Fist' if i==3 else "")

plt.legend()
plt.title("Hand Gesture Classification (Demo)")
plt.show()
```
