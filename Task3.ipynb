```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Simple dataset (fake features for demo)
# 0 = Cat, 1 = Dog
X = [
    [1, 2], [2, 3], [3, 3],   # Cats
    [6, 5], [7, 7], [8, 6]    # Dogs
]

y = [0, 0, 0, 1, 1, 1]

# Train model
model = svm.SVC()
model.fit(X, y)

# Prediction
prediction = model.predict([[5, 5]])
print("Prediction (0=Cat, 1=Dog):", prediction[0])

# Plot
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='blue', label='Cat' if i==0 else "")
    else:
        plt.scatter(X[i][0], X[i][1], color='red', label='Dog' if i==3 else "")

plt.legend()
plt.title("SVM Classification (Cat vs Dog)")
plt.show()
```
